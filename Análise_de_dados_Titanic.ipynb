```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df= pd.read_csv('/content/drive/MyDrive/passageirosDoTitanic.csv')
df
```

```python
sobrevive = df['Survived'].value_counts()
print(sobrevive)

sexo = df['Sex'].value_counts()
print(sexo)

idade = df['Age'].value_counts()
print(idade)

classe = df['Pclass'].value_counts()
print(classe)

irms = df['SibSp'].value_counts()
print(irms)

paiefilho = df['Parch'].value_counts()
print(paiefilho)

tarifa = df['Fare'].value_counts()
print(tarifa)

embarque = df['Embarked'].value_counts()
print(embarque)
```

```python
sns.histplot(df, x="Pclass", hue="Survived", multiple="dodge", bins=5, palette=["red", "green","blue"])
plt.title("Gráfico de sobreviventes por classe")
plt.xlabel("Classe")
plt.ylabel("Quantidade")
plt.xticks([1, 2, 3])
plt.show()
```

```python
sns.histplot(df, x="Sex", hue="Survived", multiple="dodge", bins=5, palette=["red", "green"])
plt.title("Gráfico de sobreviventes por sexo")
plt.xlabel("Sexo")
plt.ylabel("Quantidade")
plt.show()
```

```python
sns.histplot(df, x="Age", hue="Sex", multiple="dodge", bins=20, palette=["blue", "pink"])
plt.title("Distribuição de Idade por Sexo")
plt.xlabel("Idade")
plt.ylabel("Sexo")
plt.show()
```

```python
sns.histplot(df, x="Age", hue="Survived", multiple="dodge", bins=20, palette=["red", "blue"])
plt.title("Distribuição de Idade por Sobrevivência")
plt.xlabel("Idade")
plt.ylabel("Quantidade")
plt.show()
```

```python
sns.histplot(df, x="Parch", hue="Survived", multiple="dodge", bins=10, palette=["red", "green"])
plt.title("Gráfico de sobreviventes por filhos tidos")
plt.xlabel("Quantidade de filhos")
plt.ylabel("Quantidade de sobreviventes")
plt.show()
```

```python
df_mulheres= df[df['Sex']== 'female']

sns.histplot(df_mulheres, x="Age", hue="Survived", multiple="dodge", bins=10, palette=["red", "green"])
plt.title("Gráfico de mulheres sobreviventes")
plt.xlabel("Idade")
plt.ylabel("Quantidade de sobreviventes")
plt.show()
```

```python
df_homens= df[df['Sex']== 'male']

sns.histplot(df_homens, x="Age", hue="Survived", multiple="dodge", bins=10, palette=["red", "green"])
plt.title("Gráfico de homens sobreviventes")
plt.xlabel("Idade")
plt.ylabel("Quantidade de sobreviventes")
plt.show()
```

```python
df_homens= df[df['Sex']== 'male']

sns.histplot(df_homens, x="Pclass", hue="Survived", multiple="dodge", bins=10, palette=["red", "green"])
plt.title("Gráfico de homens sobreviventes")
plt.xlabel("Classe")
plt.ylabel("Quantidade de sobreviventes")
plt.xticks([1, 2, 3])
plt.show()
```
